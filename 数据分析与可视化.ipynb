```python
##绘制特征相关热力图
import seaborn as sns
import matplotlib.pyplot as plt
"""
需要二维表格数据，类似 Excel：
每一行：一个样本
每一列：一个数值型特征（比如：温度、速度、成本、时间、产量……）
"""
sorted_table = sorted.pivot_table(
    "销量(千克)",index="销售日期",columns="分类名称")

# 计算相关系数矩阵
corr_matrix = sorted_table.corr()

sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',center=0)
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['figure.dpi'] = 120 
plt.title('特征相关热力图')
plt.tight_layout()
plt.show()
#特征相关系数热力图de作用：不同特征之间的相关性，筛选关键特征。
```

```python
##聚类分析
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score


#画轮廓系数图确定最优聚类数k(有时间的话，可以自己在图上标注最优k点)
def silhouette_method(X, max_k=10):
    """轮廓系数：计算2~max_k的轮廓系数，返回系数列表"""
    sil_score_list = []
    for k in range(2, max_k+1):  # 轮廓系数K≥2才有效
        kmeans = KMeans(n_clusters=k, random_state=42,n_init=100)
        y_pred = kmeans.fit_predict(X)
        sil_score = silhouette_score(X, y_pred)  # 计算平均轮廓系数
        sil_score_list.append(sil_score)
    # 可视化轮廓系数曲线
    plt.rcParams['font.sans-serif'] = ['SimHei']
    plt.rcParams['figure.dpi'] = 120 
    plt.figure(figsize=(12, 5))
    plt.plot(range(2, max_k+1), sil_score_list, 
             marker='o', markersize=8, markeredgecolor='white', 
             markeredgewidth=2,linewidth=2.5, color='#2E86AB' )
    
    plt.title('轮廓系数法找最优聚类数K', fontsize=12)
    plt.xlabel('聚类数K', fontsize=10)
    plt.ylabel('平均轮廓系数', fontsize=10)
    plt.xticks(range(2,max_k+1),fontsize=10,rotation=45)
    plt.yticks(fontsize=10)
    plt.grid(alpha=0.3,linestyle="--")
    plt.tight_layout()
    plt.show()
    return sil_score_list     #返回平均轮廓系数的表



#画肘部法则图确定最优聚类数k
def elbow_method(X, max_k=10):
    """肘部法则：计算不同K的SSE，返回SSE列表"""
    sse_list = []
    for k in range(2,max_k+1):
        kmeans = KMeans(n_clusters=k, random_state=42,n_init=100)
        kmeans.fit(X)
        sse_list.append(kmeans.inertia_)  # inertia_ = SSE（误差平方和）
    # 可视化肘部图
    plt.figure(figsize=(12, 5))
    plt.rcParams['font.sans-serif'] = ['SimHei']
    plt.rcParams['figure.dpi'] = 120 
    plt.plot(range(2, max_k+1), sse_list, 
             marker='o', markersize=8, markeredgecolor='white', 
             markeredgewidth=2,linewidth=2.5, color='#2E86AB' )
    
    plt.title('肘部法则（找最优K）', fontsize=12)
    plt.xlabel('聚类数K', fontsize=10)
    plt.ylabel('SSE（误差平方和）', fontsize=10)
    plt.xticks(range(2,max_k+1),fontsize=10,rotation=45)
    plt.yticks(fontsize=10)
    plt.grid(alpha=0.3,linestyle="--")
    return sse_list  #返回惯性





#1.数据预处理
"""
需要二维表格数据，类似 Excel：
每一行：一个样本（模型会将样本分为k类)
每一列：一个数值型特征（比如：温度、速度、成本、时间、产量……）
"""
X = sorted.pivot_table(
    "销量(千克)",columns="销售日期",index="分类名称")
#2.打印相关数
sil_score_list = silhouette_method(X, max_k=5) #确定max_k
sse_list = elbow_method(X, max_k=5)
print("平均轮廓系数：",[f"{x:.2f}" for x in sil_score_list])
print("各种K值对应的惯性：",[f"{x:.2f}" for x in sse_list])

#3.K-means聚类
k=3   #确定k
kmeans = KMeans(n_clusters=k)
y_pred = kmeans.fit_predict(X)
print(y_pred)
```
